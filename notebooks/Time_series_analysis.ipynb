```python
# Jupyter Notebook to run the script and visualize the results
import subprocess
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Run the Time Series Analysis Script
subprocess.run(['python', 'Time_series_analysis.py'])

# Step 2: Load the data for visualization
df = pd.read_csv('C:/Users/user/Desktop/Github/Oilprice_analysis/data/BrentOilPrices.csv', parse_dates=['Date'], index_col='Date')

# Step 3: Plotting the Oil Prices
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Price'], label='Brent Oil Prices')  # Updated to reflect the new column name
plt.title('Brent Oil Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```
