```python
# Brent Oil Prices Analysis Notebook

# Import necessary libraries
import subprocess
import json
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Define the path to the script
script_path = 'C:/Users/user/Desktop/Github/Oilprice_analysis/scripts/Time_series_analysis.py'

# Call the script using subprocess
subprocess.run(['python', script_path])

# Load the results from the JSON file
results_path = 'C:/Users/user/Desktop/Github/Oilprice_analysis/results.json'
with open(results_path, 'r') as f:
    results = json.load(f)


```

```python
# Display ADF Test Results
print(f"ADF Statistic: {results['ADF Statistic']}")
print(f"p-value: {results['p-value']}")
print("\nARIMA Model Summary:")
print(results['ARIMA Summary'])
```

```python
# Load the Brent Oil Prices data for visualization
file_path = 'C:/Users/user/Desktop/Github/Oilprice_analysis/data/BrentOilPrices.csv'
df = pd.read_csv(file_path, parse_dates=['Date'], index_col='Date')

# Plot the historical Brent Oil Prices
plt.figure(figsize=(14, 7))
plt.plot(df['Price'], label='Brent Oil Prices')
plt.title('Historical Brent Oil Prices')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.legend()
plt.grid()
plt.show()
```
